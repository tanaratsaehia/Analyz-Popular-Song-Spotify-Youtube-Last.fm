# merge dataframe

```python
spotify = pd.read_csv('/content/cleaned_data_spotify.csv')
youtube = pd.read_csv('/content/cleaned_data_yt.csv')
last_fm = pd.read_csv('/content/cleaned_data_last_fm.csv')
```

```python
df_fm = last_fm.copy()
df_sp = spotify.copy()
df_yt = youtube.copy()

df_fm.rename(columns={'Track': 'track_name'}, inplace=True)
df_yt.rename(columns={'song_name': 'track_name'}, inplace=True)
df_sp.rename(columns={'track_name': 'track_name'}, inplace=True)

#select column
df_yt = df_yt[['track_name', 'views', 'likes']]
df_sp = df_sp[['track_name', 'popularity', 'artist', 'album', 'release_date']]
df_fm = df_fm[['track_name', 'Listeners', 'Playcount']]

# merg_spo_yt = df_sp.merge(df_yt, on=['track_name', 'artist'], how='left')
merged_left = pd.merge(df_fm, df_sp, on='track_name', how='left')
```

```python
merged_left.info()
```

```python
merged_left[merged_left['album'].isnull()]
```

```python
merged_left.loc[3079,'album'] = 'NA'
```

```python
merged_final = pd.merge(merged_left, df_yt, on='track_name', how='left')
```

```python
merged_cleaned = merged_final.dropna(subset=['views', 'likes'])
```

```python
merged_cleaned.info()
```

```python

for i in merged_cleaned.index:
  release_date = merged_cleaned.loc[i,'release_date']
  temp = release_date.split('-')
  if len(release_date) == 4:
    merged_cleaned.loc[i,'year'] = temp[0]
    merged_cleaned.loc[i,'month'] = np.NaN
    merged_cleaned.loc[i,'day'] = np.NaN
  elif len(release_date) == 7:
    merged_cleaned.loc[i,'year'] = temp[0]
    merged_cleaned.loc[i,'month'] = temp[1]
    merged_cleaned.loc[i,'day'] = np.NaN
  else:
    merged_cleaned.loc[i,'year'] = temp[0]
    merged_cleaned.loc[i,'month'] = temp[1]
    merged_cleaned.loc[i,'day'] = temp[2]
```
